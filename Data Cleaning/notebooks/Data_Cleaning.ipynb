```python
# import the dataset
df = pd.read_csv(r"D:\My Project\Visualizing US Natural Disaster Declarations\Data\raw\disaster_declarations_summaries.csv")
```

```python
# checks the data types of the columns in the dataset
display(df.dtypes)
# checks the number of unique values in each column of the dataset
display(df.nunique())
```

```python
# Parse Dates and Timezones
#Convert date columns to datetime: FEMA dates often load as strings. Use parse_dates in read_csv or pd.to_datetime.
date_columns = ['declarationDate', 'incidentBeginDate', 'incidentEndDate','disasterCloseoutDate', 'lastRefresh']
for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce')
# Check for timezone information: FEMA data may not include timezones, but if it does, ensure consistency.
# If timezone info is present, convert to a common timezone (e.g., UTC)
# df['declarationDate'] = df['declarationDate'].dt.tz_convert('UTC')  # Uncomment if timezone info exists
```

```python
# information about the dataset
display(df.info())
```

```python
#Standardize Categorical Fields
# Clean text fields like incidentType and state. Trim whitespace, fix capitalization, and unify similar entries. For instance, remove trailing spaces and use consistent casing.
df['incidentType'] = df['incidentType'].str.strip().str.title()
df['state'] = df['state'].str.strip().str.upper()
```

```python
# normalize_columns.py cell
# Normalize column names to snake_case for ease of use
def to_snake(s):
    return (
        s.strip()
        .replace(" ", "_")
        .replace("-", "_")
        .replace("/", "_")
        .lower()
    )

df.columns = [to_snake(c) for c in df.columns]
print("Normalized columns.")
display(df.columns.tolist())
```

```python
# Handle Missing Values
# Identify columns with missing values and decide on a strategy (e.g., imputation, removal
missing_values = df.isnull().sum()
display(missing_values)
```

```python
# handle missing values in the 'incidentEndDate' column by filling them with the 'incidentBeginDate' values
date_cols = [
    "declarationdate",
    "incidentbegindate",
    "incidentenddate",
    "disastercloseoutdate",
    "lastiafilingdate",
    "lastrefresh",
]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors="coerce")
```

```python
# fill missing values in the declarationtype column with the most frequent value
df["designatedIncidenttypes"] = df["designatedincidenttypes"].fillna("Not Specified")
```

```python
# verify that there are no more missing values in the dataset
missing_values_after = df.isnull().sum()
display(missing_values_after)
```

```python
# save the cleaned dataset to a new csv file
output_path = r"D:\My Project\Visualizing US Natural Disaster Declarations\Data\processed\disaster_declarations_summaries_cleaned.csv"
df.to_csv(output_path, index=False)
```
